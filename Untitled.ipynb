## This code generates a PSD from Feng's original model data. Nothing is filtered out from the PSD at this point it is just raw data. We can notice something at the Gamma band but hard to see without filtering

```python
data = np.loadtxt("LFP_elec_0")

freqs,spectrum = welch(data,fs=1000,nfft=1024)

plt.plot(freqs,spectrum)
plt.title("PSD of Feng's Data")
plt.xlabel("Hz")
plt.ylabel("Power")
plt.xlim([5,250])
plt.show()
```

# we can plot our aperiodic component from our data but first we need to put it in linear space

```python
fm = FOOOF(aperiodic_mode='knee')
fm.fit(freqs, spectrum)

ap_fit = fm._ap_fit
plt.plot(freqs,spectrum)
plt.plot(10**ap_fit)
plt.title("Data with 1/f ontop")
plt.xlabel("Hz")
plt.ylabel("Power")
plt.xlim([5,250])
plt.show()
```

# When we apply Fooof to the data and remove the 1/f component from the spectrum we can see a fairly flat graph, but for the large bump around 60Hz which is our gamma band.

```python
residual_spec = spectrum[1:] - 10**ap_fit # making arrays match

plt.plot(residual_spec)
plt.xlabel("Hz")
plt.ylabel("Power")
plt.xlim([5,250])
plt.title('residual from spectrum - aperiodic component')
plt.show()
```
